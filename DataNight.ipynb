Clean the dataset for columbs

```python
#dataset me jo columb install usmy , aur  + k sign ko replace krna
#ku k jo numeric columbs thy usy wo data set me show ni kiye ku k unky sath values thi jese koi sign thy aur strings wgera
df['Installs'] = df['Installs'].str.replace(',','')
df['Installs'] = df['Installs'].str.replace('+','')
```

```python
#size colmb me unique values dekhny k liye
df['Size'].unique()
```

```python
# Access the values of a particular column
column_values = df['Size']

# Print all the unique values in the column
unique_values = column_values.unique()
for value in unique_values:
    print(value)
```

```python
df.isnull().sum()
```

```python
df['Size']=df['Size'].str.replace('G', '000000000')
df['Size']=df['Size'].str.replace('G', '000000')
df['Size']=df['Size'].str.replace('G', '000')
```

```python
df.dtypes
```

```python
#typecasting kr k size ki value ko float me convert krna
df['Size'] = df['Size'].astype(float)
```

```python
#typecasting
df['Size'] = pd.to_numeric(df['Size'], errors= 'coerce')
```

```python
#to find the missing values
df.isnull().sum()
```

```python
#Size me missing values ko mean se replace kry gy
df['Size'] = df['Size'].fillna(value=df['Size'].mean())
```

```python
#dataset k end me ek new columb bnany k liye
df['size_mb'] = 0
```

```python
df['size_mb'] = df['Size'].apply(lambda x: x/1000000000)
```
